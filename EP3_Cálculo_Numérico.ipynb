#Exerc√≠cio 1

```python
def f(x):
  return np.int32(5-5*x**4)

def fD(x):
  return 5-5*x**4

def integral(x):
  return np.int32(5*x-x**5)
```

##a)

```python
I = 4
intervalos = []
integrais = []
erro = []

for p in range(1,11):
  N = 2**p
  h = np.float32(1/N)

  divisions = N + 1
  pontos = []
  for i in range(divisions + 1):
    pontos.append( np.float32( i*h ) )
  pontos = np.asarray(pontos)

  S = f(pontos[0]) + f(pontos[-1])
  s_impar = 0
  for i in range(2, N+1, 2):
    s_impar = s_impar + f(pontos[i])
  s_impar = 4*s_impar

  s_par = 0
  for i in range(3, N+1, 2):
    s_par = s_par + f(pontos[i])
  s_par = 2*s_par

  S = h*(S + s_impar + s_par)/3

  intervalos.append(N)
  integrais.append(S)
  erro.append(abs(I - S))
```

```python
df = pd.DataFrame({'p': range(1,11), 'N': intervalos, 'I_num': integrais, 'error': erro})
display(df)
```

#Exerc√≠cio 2

```python
def f(x, k):
  return 1 / np.sqrt(1 - k * (np.sin(x))**2)
```

```python
N = 500
H = np.pi/N
thetas = []
for i in range(N):
  thetas.append(H * i)

n_trapezios = 50
h = (np.pi/2)/n_trapezios
pontos = []
for j in range(n_trapezios + 1):
  pontos.append(h * j)

tempos = []
for theta_zero in thetas:
  k_quadrado = (1 - np.cos(theta_zero))/2
  T = f(pontos[0], k_quadrado) + f(pontos[-1], k_quadrado)
  t = 0
  for ponto in range(1,n_trapezios):
    t += f(pontos[ponto], k_quadrado)

  T = h*(T + 2*t)/2
  tempos.append(T)

tempos = np.asarray(tempos)

print(f'Existem {N} n√∫meros no intervalo de [0,pi), a divis√£o entre os √¢ngulos √© de aproximadamente {H:.4}.')
print('')
print(f'Foram usandos {n_trapezios} trap√©zios para calcular a integral, portando a dist√¢ncia entre um ponto e outro √© {h:.4}.')
```

```python
frac_t = 2*tempos/np.pi

df_2 = pd.DataFrame({'Œ∏0 (rad)': thetas, 'T (s)': tempos,'T/T_Galileu': frac_t})
display(df_2)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df_2['Œ∏0 (rad)'], df_2['T/T_Galileu'])
plt.xlabel('$Œ∏_0 (rad)$')
plt.ylabel('$T/T_{Galileu}$')
plt.axvline(x = np.pi, color = 'purple', label = 'pi', alpha = 0.5)
plt.title(f'Pendulum Period vs. Initial Angle ({N} points)')
plt.grid(True)
plt.show()
```

#Exerc√≠cio 3

```python
def yfunction(x):
  return 1 - x**2
```

##a)

```python
def random(Z_i):
  a = 1664525
  c = 1013904223
  m = 2147483647

  Z_next = (a*Z_i + c) % m
  U_i = Z_next/m
  return Z_next, U_i
```

```python
Z_0 = 13687220
N = int(input('Digite o n√∫mero de pontos: '))
print('')

for i in range(N):
  Z_0, U_i = random(Z_0)
  print(f'{i+1}¬∫ valor aleat√≥rio: {U_i}')
```

##b)

```python
def monte_carlo(Z_0):
  i = 0
  tentativas = 100

  for j in range(tentativas):
    Z_0, x = random(Z_0)
    Z_0, y = random(Z_0)
    if y <= yfunction(x):
      i += 1
    else:
      pass
  return i/tentativas, Z_0

Z_0 = 13687220
I, Z_0 = monte_carlo(Z_0 = Z_0)
print(f'Com 100 pontos a integral pelo met√≥do de Monte-Carlo √© I = {I}.')
```

##c)

```python
mintegrais = []
sintegrais = []
smintegrais = []
qintegrais = []

for n in range(1,18):
  N = 2**n
  integrais = []
  Z_0 = 13687220
  for tentativa in range(N):
    I, Z_0 = monte_carlo(Z_0 = Z_0)
    integrais.append(I)
  qintegrais.append(N)
  mintegrais.append(np.mean(integrais))
  sintegrais.append(np.std(integrais))
  smintegrais.append(np.std(integrais) / np.sqrt(N))

df_3 = pd.DataFrame({'N_t' : qintegrais,
                     'I_m': mintegrais,
                     'sigma': sintegrais,
                     'sigma_m': smintegrais})
print(df_3.tail())
```
