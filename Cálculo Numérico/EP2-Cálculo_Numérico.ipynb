```python
def eliminacao_gaussn(M):
  i = 1
  for l in range(len(M)-1): #a l-ésima linha é a que vai multiplicada e somada/subtraída as demais
    for linha in range(len(M)-i):
      M[i:][linha] = M[i:][linha] - M[l] * (M[i:][linha][l]/M[l][l]) #M[i:] pega os elementos da matriz que estão logo abaixo da nossa l-ésima linha, no caso 'linha' é exatamente a linha que vai ser somada a esta.
      print(M)
    i += 1
  return M
```

```python
def solucaon(M):
  x_final = M[-1][-1]/M[-1][-2] #já resolve o sistema da última linha, pois é necessário para calcular as demais
  sol = [[x_final]] #lista das soluções
  linha = -1
  while len(sol) < len(M):
    ab = 0 #ab é soma de todos os termos a (coeficientes) com os b(soluções já encontradas)
    for coef in range(len(sol)):
      ab += M[linha-1][-coef-2] * sol[coef][0]
    x_n = (M[linha-1][-1] - ab)/M[linha-1][linha-2] # M[linha-1][-1] é o resultado da linha analisada e M[linha-1][linha-2] é o coeficiente
    sol.append([x_n])
    linha -= 1
  return sol[::-1]
```

#Questão c)

```python
def calculo_jacobi(matriz, solucao_i):
  x_k = []
  for k in range(len(matriz)):
    linha = np.delete(matriz[k], k)
    linha = np.delete(linha, -1)
    vetor_solucao_k = np.delete(solucao_i, k)
    solucao = ( matriz[k][-1] - np.dot(np.asarray(linha), np.asarray(vetor_solucao_k)) ) / matriz[k][k]
    x_k.append(solucao)
  return x_k
```

```python
def metodojacobi(matriz, chute, epsilon):
  solucao_i = chute
  solucoes = [solucao_i]
  erros = []
  k = 0
  K = []
  while abs( max(np.asarray(solucoes[-1]) - np.asarray(calculo_jacobi(matriz, solucao_i)) ) ) > epsilon:
    erros.append(abs( max(np.asarray(solucoes[-1]) - np.asarray(calculo_jacobi(matriz, solucao_i)) ) ))
    solucao_i = calculo_jacobi(matriz, solucao_i)
    solucoes.append(solucao_i)
    K.append(k)
    k += 1
  erros.append(abs( max(np.asarray(solucoes[-1]) - np.asarray(calculo_jacobi(matriz, solucao_i)) ) ))
  return solucoes, erros, k
```

```python
solucao_J, erros, k =  metodojacobi(Jmatriz, X, epsilon)
```

```python
def calculo_GS(matriz, solucao_i):
  nova_solucao = solucao_i.copy()
  for k in range(len(matriz)):
    linha = np.delete(matriz[k], k)
    linha = np.delete(linha, -1)
    vetor_solucao_k = np.delete(nova_solucao, k)
    nova_solucao[k] = ( matriz[k][-1] - np.dot(np.asarray(linha), np.asarray(vetor_solucao_k)) ) / matriz[k][k] #atualiza o k-ésimo termo da lista de solução.
  return nova_solucao
```

```python
def metodoGS(matriz, chute):
  solucao_i = chute
  solucoes = [solucao_i]
  erros = []
  k = 0
  K = []
  while abs( max(np.asarray(solucoes[-1]) - np.asarray(calculo_GS(matriz, solucao_i)) ) ) > epsilon:
    erros.append(abs( max(np.asarray(solucoes[-1]) - np.asarray(calculo_GS(matriz, solucao_i)) ) ))
    solucao_i = calculo_GS(matriz, solucao_i)
    solucoes.append(solucao_i)
    K.append(k)
    k += 1
  erros.append(abs( max(np.asarray(solucoes[-1]) - np.asarray(calculo_GS(matriz, solucao_i)) ) ))
  return solucoes, erros, k
```

```python
X = []
print("O vetor chute de solução é composto por: ")
for i in range(3):
  x = float(input(f'x_{i+1} = '))
  X.append(x)
X = np.array(X)
```

```python
solucao_GS, erros, k = metodoGS(Jmatriz, X)
```

```python
#Gemini programou


data = {
    'Iteração': range(len(solucao_GS)),
    'Erro': erros,
    'Solução': solucao_GS
}

df_GS = pd.DataFrame(data)

# Use the solution data from the Gauss-Seidel method (solucao_GS)
solucao_df = pd.DataFrame(df_GS['Solução'].tolist(), index=df_GS.index)

# Rename the new columns
solucao_df.columns = [f'I_{i+1}' for i in range(solucao_df.shape[1])]

# Concatenate the new columns with the original DataFrame (excluding the original 'Solução' column)
df_GS_expanded = pd.concat([df_GS.drop('Solução', axis=1), solucao_df], axis=1)

display(df_GS_expanded)
```
