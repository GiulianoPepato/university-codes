#Questão a)

```python
import numpy as np
```

```python
def function(x):
  return x**3 + x + np.cos(x)
```

```python
x_1, x_2, epsilon, i = -0, -1, 0.00001, 1
x_1list = []
x_2list = []
erros = []
x_mlist = []
fx = []
while abs(x_2 - x_1) > epsilon:
  e_np1 = 1/2**i
  x_1list.append(x_1)
  x_2list.append(x_2)
  erros.append(e_np1)
  x_m = (x_1 + x_2)/2
  x_mlist.append(x_m)
  i += 1
  if function(x_m) < 0:
    x_2 = x_m
  else:
    x_1 = x_m
  fx.append(function(x_m))
```

```python
import pandas as pd
```

```python
df = pd.DataFrame({
    'x_1': x_1list,
    'x_2': x_2list,
    'x_m': x_mlist,
    'f(x_m)': fx,
    'e_n': erros
})
```

```python
df
```

#Questão b)

```python
def g(x):
  return x - function(x)/(3*x**2 + 1 - np.sin(x))

def erro_r(x, f):
  return abs( (f - x)/ x )
```

```python
x_n = 0.01
X = [x_n]
F_b = [function(x_n)]
erros = [erro_r(x_n, g(x_n))]
while erro_r(x_n, g(x_n)) > epsilon:
  x_n = g(x_n)
  X.append(x_n)
  F_b.append(function(x_n))
  erros.append(erro_r(x_n, g(x_n)))
```

```python
df_b = pd.DataFrame({
    "x": X,
    "f(x)": F_b,
    "e_n": erros
})
```

```python
def energy(E):
  return np.tan( np.sqrt( E+9 ) ) -  np.sqrt(-E/(E+9))

def xnmaisum(x, y):
  return x -  energy(x)*(x-y)/(energy(x)-energy(y))
```

```python
x_n = -7.5
x_n1 = -7.2
X = [x_n]
X_1 = [x_n1]
fx =[energy(x_n)]
fx1 = [energy(x_n1)]
erros = [abs(xnmaisum(x_n, x_n1) - x_n)]
#erro_r(x_n, xnmaisum(x_n, x_n1))
while abs(xnmaisum(x_n, x_n1) - x_n) > epsilon:
  x_np1 = xnmaisum(x_n, x_n1)
  x_n1 = x_n
  x_n = x_np1
  X.append(x_n)
  X_1.append(x_n1)
  fx.append(energy(x_n))
  fx1.append(energy(x_n1))
  erros.append(abs(xnmaisum(x_n, x_n1) - x_n))
```
